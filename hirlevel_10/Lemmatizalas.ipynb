```python
#https://github.com/LibreOffice/dictionaries/tree/master/hu_HU
import hunspell
# Hunspell inicializálás
h = hunspell.HunSpell(
    "hu_HU.dic",
    "hu_HU.aff"
)
```

```python

import re

# kikapcsolva
# normalizáló, ami a Hunspell által visszaadott szóalakot igetővé próbálja alakítani.
# infinitívusz (-ni végű), ikes igék
'''
def normalize_hunspell_stem(stem):
    # infinitívusz → igető
    if stem.endswith("ni"):
        return stem[:-2]

    # ikes igék (történik, látszik, tetszik stb.)
    if stem.endswith("ik"):
        return stem[:-2]

    return stem
'''


# lemmatizálás, input: szó
def lemmatize_word(word):
    clean = re.sub(r'^\W+|\W+$', '', word, flags=re.UNICODE)
    stems = h.stem(clean)
    if stems:
        return stems[0].decode("utf8")
        #foo = stems[0].decode("utf8")
        #return normalize_hunspell_stem(foo)
    return word
# lemmatizálás , input: mondat
def lemmatize_text(text):
    words = text.split()
    return " ".join(lemmatize_word(w) for w in words)
```

```python
eredeti = "gyerekek"
modositott = lemmatize_word(eredeti)
print(modositott)
```

```python
eredeti = "tömések"
modositott = lemmatize_word(eredeti)
print(modositott)
```

```python
szoveg = """
TÖMÉSEK
A fogtömés szuvas fogak helyreállítására szolgál.
Esztétikus, fogszínű tömések érhetők el.
A kezelés megakadályozza a további romlást.
Gyors és hatékony megoldás.
Helyi érzéstelenítésben történik.

"""
```

```python
eredeti = szoveg
modositott = lemmatize_text(eredeti)
print(modositott)
```
