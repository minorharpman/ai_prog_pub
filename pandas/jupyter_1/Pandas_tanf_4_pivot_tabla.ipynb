## Pandas: pivot tábla

```python
# Importáljuk a pandas könyvtárat, amit általában "pd" néven szokás rövidíteni.
# A pandas egy Python könyvtár táblázatos (Excel-szerű) adatok kezelésére.
import pandas as pd
```

```python
# Beolvasunk egy Excel fájlt DataFrame-be.
# A pd.read_excel() függvény egy Excel fájlt olvas be és DataFrame-et ad vissza.
# A DataFrame a pandas egyik legfontosabb adattípusa, egy táblázatszerű adatstruktúra.
df = pd.read_excel("mobil_bob_shop.xlsx")
```

```python
# Megnézzük az első néhány sort, hogy lássuk, hogyan néz ki az adatkeret.
df.head(3)
```

```python
# Új oszlopot hozunk létre 'bevétel' néven.
# A df['mennyiség'] és df['ár'] oszlopokat összeszorozzuk soronként.
# Ez szintén egy pandas Series típusú objektum, amit új oszlopként tárolunk.
df['Bevétel'] = df['Mennyiség'] * df['Ár']
df
```

## Pivot tábla készítése a bevételekről kategória és régió szerint

pivot tábla (forgatott tábla, kereszt tábla) táblázatos adatösszesítő eszköz,  táblák strukturált összesítésére használnak



* A pivot tábla (angolul pivot table) a Pandas könyvtárban egy adatátalakítási és összegzési eszköz, amely lehetővé teszi, hogy egy DataFrame-ből könnyen áttekinthető, aggregált (összesített) formában nyerjünk ki adatokat.
* Mire jó a pivot tábla?
** Adatok csoportosítása
** Összesítések


Példa: Pivot tábla készítése a bevételekről kategória és régió szerint

* pd.pivot_table(data, values=None, index=None, columns=None, aggfunc='sum')
*  pivot_table() a pandas egyik legfontosabb függvénye csoportosított, tömörített táblák készítéséhez

* Paraméterek magyarázata:
*- values= mely oszlop(oka)t szeretnéd aggregálni , 'Bevétel' → ez az oszlop adja az értékeket (amit összeadunk)
* - index='Kategória' → ez lesz a sorindex (soronkénti bontás)
* - columns='Régió' → ez lesz az oszlopok bontása (pl. külön oszlop minden régióhoz)
* - aggfunc='sum' → az aggregálási függvény: összeadás (bevételösszesítés)
* pivot_df = pd.pivot_table(df, values='Bevétel', index='Kategória', columns='Régió', aggfunc='sum')
* pivot_df

```python
# Kiírjuk a kész pivot táblát – kategóriánként és régiónként összesített bevételekkel
print("Pivot tábla – Bevétel kategóriánként és régiónként:")
pivot_df = pd.pivot_table(df, values='Bevétel', index='Kategória', columns='Régió', aggfunc='sum')
pivot_df
```

```python
pivot_df2 = pd.pivot_table(df, values='Bevétel', index='Régió', columns='Kategória', aggfunc='sum')
pivot_df2
```

# Szűrés
* Oszlop (columns) Régió szűrés: Budapest, Debrecen

```python
szurt_df = df[df['Régió'].isin(['Budapest', 'Debrecen'])]
szurt_df
```

```python
pivot_df = pd.pivot_table(szurt_df, values='Bevétel', index='Kategória', columns='Régió', aggfunc='sum')
pivot_df
```

#Multi indexű pivot tábla

```python
pivot_df = pd.pivot_table(df, values='Bevétel', index=['Kategória', 'Akció'], columns='Régió', aggfunc='sum')
pivot_df
```

#Többféle aggregálás egyszerre

```python
# Többféle statisztikai mutatót is kérhetünk egy pivot táblába: például összeget és átlagot is egyszerre
szurt_df = df[df['Régió'].isin(['Budapest', 'Debrecen'])]
multi_pivot = pd.pivot_table(
    szurt_df ,
    values='Bevétel',
    index='Kategória',
    columns='Régió',
    aggfunc=['sum', 'min']  # Többféle aggregáció
)

print(" Pivot tábla – Összeg és Átlag egyszerre:")
multi_pivot
```
